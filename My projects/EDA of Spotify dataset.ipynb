```python
plt.figure(figsize=(8,3))
sns.distplot(data.Streams,
           bins=20,
           kde_kws={"lw":4,"color":"green"},
            hist_kws={"alpha":0.3,"color":"black"}
            )
plt.xlabel("weight (lbs)",fontsize=16,labelpad=15)
plt.yticks([])
plt.show()
```

```python
data.groupby("Song Name")[["Popularity","Danceability"]].sum().plot.bar(color=["red","blue"],figsize=(8,5))
plt.ylabel("Popularity vs Danceability")
plt.show()
```
